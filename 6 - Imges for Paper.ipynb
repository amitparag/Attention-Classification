```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset
import matplotlib.patches as patches
import seaborn as sns
%matplotlib inline

# Plot the loss curve
plt.rcParams["font.family"] = "Times New Roman"

sns.set_theme(style='whitegrid', palette='deep', font_scale=3, color_codes=True, rc=None)
```

```python
image1 = mpimg.imread('./assets/robot.png')
image2 = mpimg.imread('./assets/sensor.png')
image3 = mpimg.imread('./assets/marker.png')
```

```python
# Create a 1x3 grid of subplots
plt.figure(figsize=(12, 4))

# Plot the first image in the first subplot
plt.subplot(131)
plt.imshow(image1)
plt.title('Image 1')
plt.axis('off')

# Plot the second image in the second subplot
plt.subplot(132)
plt.imshow(image2)
plt.title('Image 2')
plt.axis('off')

# Plot the third image in the third subplot
plt.subplot(133)
plt.imshow(image3)
plt.title('Image 3')
plt.axis('off')

# Adjust spacing between subplots
plt.tight_layout()

# Show the grid of images
plt.show()
```

```python
# Create a figure with a specific size
plt.figure(figsize=(19.2, 10.8))

# Plot Image 1 in the left half of the grid
plt.subplot(1, 2, 1)
plt.imshow(image1)
plt.title('Franka Panda Arm', pad=20)  # Adjust the 'pad' value for padding
plt.axis('off')

# Create a 2x1 grid for the right half of the grid
plt.subplot(2, 2, 2)
plt.imshow(image2)
plt.title('Gelsight Sensors', pad=20)  # Adjust the 'pad' value for padding
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(image3)
plt.title('Data from Sensors', pad=20)  # Adjust the 'pad' value for padding
plt.axis('off')

# Adjust spacing between subplots
plt.tight_layout()

# Save the figure to a file
plt.savefig('./assets/setup.png', dpi=180)

# Show the grid of images
plt.show()
```
